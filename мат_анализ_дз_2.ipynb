## №3 На языке Python предложить алгоритм, вычисляющий численно предел 
## lim n/n!^(1/n) с точностью 10^-7

```python
n = 1
s2 = 1
for i in range(10**7-1):
    s = s2
    n += 1
    s2 = n**(1/n) * s**((n-1)/n)
    if n%(5 * 10**5) == 0:
        print(1/(s2-s))
print()
1/(s2 - s), n
```
