# Тема "Множество"

Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

```python
a = {1, 2, 8, 9}
b = {3, 7, 8}
c = {1, 3, 4, 5, 9}
```

## Объединение

```python
aub = a.union(b)
aub
```

```python
auc = a.union(c)
auc
```

```python
buc = b.union(c)
buc
```

```python
aubuc = a.union(b).union(c)
aubuc
```

## Пересечение

```python
aib = a.intersection(b)
aib
```

```python
aic = a.intersection(c)
aic
```

```python
bic = b.intersection(c)
bic
```

```python
aibic = a.intersection(b).intersection(c)
aibic
```

## Разность

```python
adb = a.difference(b)
adb
```

```python
bda = b.difference(a)
bda
```

```python
adc = a.difference(c)
adc
```

```python
cda = c.difference(a)
cda
```

```python
bdc = b.difference(c)
bdc
```

```python
cdb = c.difference(b)
cdb
```

```python
adbdc = a.difference(b).difference(c)
adbdc
```

```python
bdadc = b.difference(a).difference(c)
bdadc
```

```python
cdadb = c.difference(a).difference(b)
cdadb
```

## Симметрическая разность

```python
asdb = b.symmetric_difference(a)
asdb
```

```python
asdc = a.symmetric_difference(c)
asdc
```

```python
bsdc = b.symmetric_difference(c)
bsdc
```

```python
asdbsdc = a.symmetric_difference(b).symmetric_difference(c)
asdbsdc
```
